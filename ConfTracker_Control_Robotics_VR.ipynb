```python
# Importing libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests


#Fetch Pages
URL = "https://ifac.papercept.net/conferences/scripts/start.pl/robots.txt"
page1 = requests.get(URL)
URL2="https://css.paperplaza.net/conferences/scripts/start.pl/robot.txt"
page2 = requests.get(URL2)
URL3="https://controls.papercept.net/conferences/scripts/start.pl/robot.txt"
page3 = requests.get(URL3)
URL4="https://ras.papercept.net/conferences/scripts/start.pl/robot.txt"
page4 = requests.get(URL4)
URL5="https://its.papercept.net/conferences/scripts/start.pl/robot.txt"
page5 = requests.get(URL5)


#Parse(Extract) HTML
page1_vals= BeautifulSoup(page1.content, "html.parser")
page2_vals= BeautifulSoup(page2.content, "html.parser")
page3_vals= BeautifulSoup(page3.content, "html.parser")
page4_vals= BeautifulSoup(page4.content, "html.parser")
page5_vals= BeautifulSoup(page4.content, "html.parser")


#Update links
f = open("Links.txt", "w")


for link in page1_vals.find_all("a"):
   data = link.get('href')
   if data !=None:
    f.write(data)
    f.write("\n")


for link in page2_vals.find_all("a"):
   data = link.get('href')
   if data !=None:
    f.write(data)
    f.write("\n")


for link in page3_vals.find_all("a"):
   data = link.get('href')
   if data !=None:
    f.write(data)
    f.write("\n")


for link in page4_vals.find_all("a"):
   data = link.get('href')
   if data !=None:
    f.write(data)
    f.write("\n")

for link in page5_vals.find_all("a"):
   data = link.get('href')
   if data !=None:
    f.write(data)
    f.write("\n")

f.close()


#Update Details
f1=open("InfoConf.txt","w",encoding="utf-8")


data=page1_vals.find_all("tr")
for i in data:
    ##print(i.prettify(), end="\n"*2)
    title_element = i.find("a")
    if title_element!=None:
        f1.write(title_element.text.strip())
        f1.write("\n")
        link=title_element.get('href')
        if link!=None:
            f1.write(link)
            f1.write("\n")
        #print(title_element.text.strip(), end="\n")
        #print(title_element.get('href'))
    date_element=i.find("div",class_="announcement")
    if date_element!=None:
        f1.write(date_element.text.strip())
        f1.write("\n")
        #print(date_element.text.strip(), end="\n")
    d1=i.find("br")
    if d1!=None:
        f1.write(d1.text.strip())
        f1.write("\n")
        #print(d1.text.strip(), end="\n"*2)


data=page2_vals.find_all("tr")
for i in data:
    ##print(i.prettify(), end="\n"*2)
    title_element = i.find("a")
    if title_element!=None:
        f1.write(title_element.text.strip())
        f1.write("\n")
        link=title_element.get('href')
        if link!=None:
            f1.write(link)
            f1.write("\n")
        #print(title_element.text.strip(), end="\n")
        #print(title_element.get('href'))
    date_element=i.find("div",class_="about")
    if date_element!=None:
        f1.write(date_element.text.strip())
        f1.write("\n")
        #print(date_element.text.strip(), end="\n")
    d1=i.find("br")
    if d1!=None:
        f1.write(d1.text.strip())
        f1.write("\n")
        #print(d1.text.strip(), end="\n"*2)


data=page3_vals.find_all("tr")
for i in data:
    ##print(i.prettify(), end="\n"*2)
    title_element = i.find("a")
    if title_element!=None:
        f1.write(title_element.text.strip())
        f1.write("\n")
        link=title_element.get('href')
        if link!=None:
            f1.write(link)
            f1.write("\n")
        #print(title_element.text.strip(), end="\n")
        #print(title_element.get('href'))
    date_element=i.find("div",class_="announcement")
    if date_element!=None:
        f1.write(date_element.text.strip())
        f1.write("\n")
        #print(date_element.text.strip(), end="\n")
    d1=i.find("br")
    if d1!=None:
        f1.write(d1.text.strip())
        f1.write("\n")
        #print(d1.text.strip(), end="\n"*2)


data=page4_vals.find_all("tr")
for i in data:
    ##print(i.prettify(), end="\n"*2)
    title_element = i.find("a")
    if title_element!=None:
        f1.write(title_element.text.strip())
        f1.write("\n")
        link=title_element.get('href')
        if link!=None:
            f1.write(link)
            f1.write("\n")
        #print(title_element.text.strip(), end="\n")
        #print(title_element.get('href'))
    date_element=i.find("div",class_="announcement")
    if date_element!=None:
        f1.write(date_element.text.strip())
        f1.write("\n")
        #print(date_element.text.strip(), end="\n")
    d1=i.find("br")
    if d1!=None:
        f1.write(d1.text.strip())
        f1.write("\n")
        #print(d1.text.strip(), end="\n"*2)

data=page5_vals.find_all("tr")
for i in data:
    ##print(i.prettify(), end="\n"*2)
    title_element = i.find("a")
    if title_element!=None:
        f1.write(title_element.text.strip())
        f1.write("\n")
        link=title_element.get('href')
        if link!=None:
            f1.write(link)
            f1.write("\n")
        #print(title_element.text.strip(), end="\n")
        #print(title_element.get('href'))
    date_element=i.find("div",class_="announcement")
    if date_element!=None:
        f1.write(date_element.text.strip())
        f1.write("\n")
        #print(date_element.text.strip(), end="\n")
    d1=i.find("br")
    if d1!=None:
        f1.write(d1.text.strip())
        f1.write("\n")
        #print(d1.text.strip(), end="\n"*2)

f1.close()


```
